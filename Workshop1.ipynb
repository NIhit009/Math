```python
def square(n):
  return n*n
def sum_squares(x):
  sum = 0
  for n in range(x):
    sum += square(n)
  return sum
print(sum_squares(10)) #Output= 285
```

```python
# Exercise - 1:
def fib():
  n = int(input("Enter the number of terms: "))
  a = 0
  b = 1
  print(a)
  print(b)
  for i in range(n - 2):
    c = a + b
    a = b
    b = c
    print(c)
fib()
```

```python
# Exercise - 2
def myPow(x, n):
  result = 1
  i = 0
  while i < n:
    result *= x
    i += 1
  return result
print(myPow(4, 4))
```

```python
# Exercise - 3
def sumDigits(x):
  sum = 0
  while x != 0:
    sum += x % 10
    x = x // 10
  print(sum)
sumDigits(234)
```
